# Problem No.4

$$ P = A\sigma \epsilon T^4 $$

```python
# Since, we can only plot a graph btw dimensionless quantities.
```

```python
#  $$ P / (A\sigma \epsilon) = T^4 $$ using Latex format, we can write.
```

$$ P / (A\sigma \epsilon) = T^4 $$

```python
# where T is also a function of time and not a Dimensionless quantity.
```

```python
# $$ T(t) = T_0 \left( \frac{1}{1+e^{-kt}} \right) $$ with this, we can write.
```

$$ T(t) = T_0 \left( \frac{1}{1+e^{-kt}} \right) $$

$$ T(t)/T_0 = \left (\frac{1}{1+e^{-kt}} \right) $$

```python
# Now, we can define a dimensionless quantity as :
```

$$  P / ( A\sigma\epsilon T_0^4 ) = \left (\frac{1}{1+e^{-kt}} \right)^4 $$

```python
kt = np.linspace(0,10,100)
p = 1/(1 + np.exp(-kt))**4
```

```python
plt.plot(kt, p)
```

```python
E = np.cumsum(p)* (kt[1]-kt[0])
```

```python
plt.plot(kt, E)
plt.title("Energy emitted by the Star")
plt.xlabel('$kt$')
plt.ylabel(r'$\left(\frac{k}{A\sigma\epsilon T_0^4}\right) E(kt)$', fontsize=20)
```

## Problem No.9

$$ E = \alpha\int |\left(\frac{d\vec{r}}{dt}\right)|dt - \int \vec{F}.\left(\frac{d\vec{r}}{dt}\right)dt $$

Given : $\alpha = 5 , \vec{r_1} = (0,0)\ and\ \vec{r_2} = (0,10)$

$\vec{F} = (0, -2/(x+1)^2)$

Find the optimal value of A in $\vec{r}(t) = Asin(\pi t/10)\hat{x} + t\hat{y} $ that minimizes the work.

### Solution :

```python
from scipy.integrate import quad
from scipy.optimize import minimize
```

```python
def f(A): 
    integrand = lambda t: 5*np.sqrt(1+(np.pi*A/10)**2 * np.cos(np.pi*t/10)**2) + 2 / (1 + np.abs(A*np.sin(np.pi*t/10)))**2
    return quad(integrand, 0, 10)[0]
```

```python
minimize(f, 0.001)
```

```python
minimize(f, 0.001).x
```
