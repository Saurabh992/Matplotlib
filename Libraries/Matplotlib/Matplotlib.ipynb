### Single Axes in one Figure

```python
fig, ax = plt.subplots(1, 1, figsize=(8,3))
ax.plot(x, y, 'o--', lw=0.7, ms=3, label='experimental')
ax.set_ylabel('Time (t)')
ax.set_xlabel('Voltage (V)')
ax.set_title('EMF')
ax.legend(loc='lower right', fontsize=12)
```

### Multiple Axes in 1 Figure

```python
fig, axes = plt.subplots(2, 2, figsize=(12,12))

ax = axes[0][0]
ax.plot(x, y, 'o--', lw=0.7, ms=3, label='experimental')
ax.set_ylabel('Time (t)')
ax.set_xlabel('Voltage (V)')
ax.legend(loc='lower right', fontsize=12)
ax.text(0.05, 0.3, 'y = Sinx', transform=ax.transAxes, fontsize=12)
ax.tick_params(axis='both', labelsize=10)

ax = axes[1][0]
ax.hist(res, bins=30, density=True, histtype='step', label='Histogram 1')
ax.legend(loc='upper right', fontsize=12)

ax = axes[0][1]
ax.plot(x, y_2, '-', lw=0.7, ms=3, label='Theoretical')
ax.set_xlabel('Voltage (V)')
ax.legend(loc='lower right', fontsize=12)

ax = axes[1][1]
ax.hist(res_2, bins=30, density=True, histtype='step', label='Histogram 2')
ax.legend(loc='upper right', fontsize=12)

plt.show()
```

### Example

```python
res_a1 = 0.2*np.random.randn(1000) + 0.4
res_a2 = 0.25*np.random.randn(1000) + 0.4
res_b1 = 0.21*np.random.randn(1000) + 0.3
res_b2 = 0.22*np.random.randn(1000) + 0.3
```

```python
parameters_1= '\n'.join(((
                        r'$\sigma_a$ = {:.4f}'.format(np.std(res_b1)),
                        r'$\sigma_b$ = {:.4f}'.format(np.std(res_b1)))))

parameters_2 = '\n'.join((
                        r"$\sigma_{a}$ = %.4f" % np.std(res_a2),
                        r"$\sigma_{b}$ = %.4f" % np.std(res_b2)))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

ax = axes[0]
ax.hist(res_a1, bins=30, density=True, histtype='step', label='A', color='blue')
ax.hist(res_b1, bins=30, density=True, histtype='step', label='B', color='red')
ax.legend(loc='upper right', fontsize=10, edgecolor='black')
ax.text(0.1, 0.8, parameters_1, transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='black'))
ax.set_ylabel(r'Density ($\rho$)')
ax.set_title('Experiment 1')

fig.text(0.5, -0.04, r'$\Delta E$ (Joules)', ha='center', fontsize=20 )

ax = axes[1]
ax.hist(res_a2, bins=30, density=True, histtype='step', label='A')
ax.hist(res_b2, bins=30, density=True, histtype='step', label='B')
ax.text(0.1, 0.8, parameters_2, transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='black'))
ax.legend(loc='upper right', fontsize=10, edgecolor='black')
ax.set_title('Experiment 2')
```

## StreamPlots

For plotting vector fields

```python
w = 3
N = np.linspace(-3, 3, 100)
x, y = np.meshgrid(N, N)
U = -1 - x**2 + y
V = 1 + x - y**2
speed = np.sqrt(U**2 + V**2)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(7,7))

ax = axes[0][0]
ax.streamplot(x, y, U, V)

ax = axes[0][1]
ax.streamplot(x, y, U, V, color=speed)

ax = axes[1][0]
lw = 5*speed / speed.max()
ax.streamplot(x, y, U, V, linewidth=lw)

ax = axes[1][1]
lw = 5*speed / speed.max()
seedpoints = np.array([[0,1], [1,0]])
ax.streamplot(x, y, U, V, start_points=seedpoints)
```

# Animation

Animation requires relating the time of your problem to the frame of the animation, then specifying the frames per second

Below is a sin wave moving at 3m/s. Translate this into an animation

```python
def f(A, k, w, x, t):
    return A*np.sin(k*x - w*t)

x = np.linspace(0, 10*np.pi, 1000)
```

```python
plt.plot(x, f(1, 1, 3, x, 0))
plt.plot(x, f(1, 1, 3, x, 1))
```

Create Animation

```python
from matplotlib import animation
from matplotlib.animation import PillowWriter
import matplotlib
matplotlib.use("Qt5Agg")
```

```python
def f(A, k, w, x, t):
    return A*np.sin(k*x - w*t)

x = np.linspace(0, 10*np.pi, 1000)

fig, ax = plt.subplots(1, 1, figsize=(8,4))
ax.set_xlim(0, 10*np.pi)
ax.set_ylim(-1.5, 1.5)
line, = ax.plot([], [])
time_text = ax.text(0.65, 0.9," ", transform=ax.transAxes, fontsize=10, bbox=dict(facecolor='white', edgecolor='black'))
```

```python
def animate(i):
    line.set_data(x, f(1, 1, 3, x, t=i/50))
    time_text.set_text(f"t = {i/50 :.2f}")
    return line, time_text

ani = animation.FuncAnimation(fig, animate, frames=240, interval=50, blit=True)
ani.save('animation.mp4', writer='ffmpeg', fps=50, dpi=100)
```
